```python
#Dictionary can be of any data type
```

```python
dict1={
    "Name":"Gaurav kamal",
    "Roll no.":10056,
    "Marks":[80,70,78,86],
    "Failed?":False
}
print(dict1)
```

```python
#updating dictionary
```

```python
dict1={"Name":"Gaurav kamal",
    "Roll no.":10056,
    "Marks":[80,70,78,86],
    "Failed?":False}
dict1.update({"Marks":[80,81,82]})
print(dict1)
```

```python
#adding new keys and values
```

```python
dict1["Phone no."] = 123456
print(dict1)
```

```python
#removing items from a dictionary
#pop method
#popitem method
#delete method
#clear method
```

```python
dict1={"Name":"Gaurav kamal",
    "Roll no.":10056,
    "Marks":[80,70,78,86],
    "Phone no.":[123456789],
    "Failed?":False}
dict1.pop("Phone no.")
print (dict1)
```

```python
del dict1['Marks']
print (dict1)
```

```python
dict1={"Name":"Gaurav kamal",
    "Roll no.":10056,
    "Marks":[80,70,78,86],
    "Failed?":False}
dict1.popitem()
print (dict1)
```

```python
dict1={"Name":"Gaurav kamal",
    "Roll no.":10056,
    "Marks":[80,70,78,86],
    "Failed?":False}
dict1.clear()
print(dict1)
```

```python
#copy dictionary function
```

```python
d2={
    "Name":"Gaurav kamal",
    "Roll no.":10056,
    "Marks":[80,70,78,86]
}
print(d2)
```

```python
dict1={
    "Name":"Gaurav kamal",
    "Roll no.":10056,
    "Marks":[80,70,78,86]
}
dict2=dict1.copy()
print(dict2)
```

```python
dict1={
    "Name":"Gaurav kamal",
    "Roll no.":10056,
    "Marks":[80,70,78,86]
}
dict2=dict(dict1)
print(dict2)
```
